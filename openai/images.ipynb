# Dall-e-3

Solo permite la creación de imágenes

```python
# Soporta tamaños de: 1024x1024, 1792x1024, or 1024x1792 para dall-e-3
# Quality: hd, statandar. Solo en dall-e-3
# response_format: url, b64_json, el url solo esta disponible por una hora
# response_format

from openai import OpenAI
client = OpenAI()

response = client.images.generate(
    model="dall-e-3",
    prompt="An avatar for an expert in Data Analysis",
    size="1024x1024",
    quality="standard",
    response_format="url",
    n=1, # numero de imagenes a generar
)

print(response.data[0].url)
print(response.data)
```

Imagen en Base64_json

```python
prompt = """
You must generate an avatar with these features:

# Features
## Background: Always use a white background
## Body: Show the entire body of the avatar gazing you
## Light: Soft and bright
## Colors: Use vivid color following best practices for disigning
## Apparel: Include additional accesories according to the avatar's description
## Position: Only draw one avatar in the middle of the image
## Style: Use cartton images

# Avatar description

"""
```

```python
# Soporta tamaños de: 1024x1024, 1792x1024, or 1024x1792 para dall-e-3
# Quality: hd, statandar. Solo en dall-e-3
# response_format: url, b64_json
# response_format

from openai import OpenAI
client = OpenAI()

response = client.images.generate(
    model="dall-e-3",
    prompt=prompt + "An avatar for an expert in Data Analysis",
    size="1024x1024",
    quality="standard",
    response_format="b64_json",
    n=1, # numero de imagenes a generar
)

print(response.data)
```

```python
import base64

# Supongamos que `response` es el resultado de la llamada a OpenAI API
b64_image = response.data[0].b64_json  # Extrae el contenido base64 de la respuesta

# Decodifica el contenido base64
image_data = base64.b64decode(b64_image)

# Guarda la imagen en un archivo
with open("avatar.png", "wb") as image_file:
    image_file.write(image_data)

print("avatar.png")
```
