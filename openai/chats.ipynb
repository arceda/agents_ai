```python
from openai import OpenAI
from dotenv import load_dotenv, find_dotenv
from pathlib import Path
import os

# en windows es necesario especificar la ruta del .env
load_dotenv()
client = OpenAI(api_key=os.environ.get("OPENAI_API_KEY"))
```

Prompt examle to generate Data Model

```python
f = open("prompts/create-data-model-v2", "r")
prompt = f.read()
f.close() 

completion = client.chat.completions.create(
  model="gpt-4o",
  messages=[
    {
        "role": "system", 
        "content": prompt
    },
    {
        "role": "user", 
        "content": "Give me a data model for a web page which will have a onboarding process for new users on the platform. The user interface should start with a welcome message and a 'Get Started' button. Clicking this button will guide the user through a series of steps: entering personal information, setting up preferences, and a brief tutorial. Each step should have an option to skip, and users can track their progress through a progress bar at the top of the screen. At the end of the onboarding, users should receive a confirmation message indicating successful completion and an option to explore the platform"
    }
  ]
)
message = completion.choices[0].message.content
message
```
